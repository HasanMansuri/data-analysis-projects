```python
import pandas as pd
import matplotlib.pyplot as plt


#  Project 1: iPhone Sales Analysis

print("\n--- Project 1: iPhone Sales Analysis ---")
iphone = pd.DataFrame({
    "Model": ["iPhone 6", "iPhone 6s", "iPhone 7", "iPhone 8", "iPhone X",
              "iPhone XR", "iPhone 11", "iPhone 12", "iPhone 13", "iPhone 14"],
    "Sales": [220, 174, 160, 125, 90, 75, 102, 100, 120, 90]
})
print(iphone.head())
print("\nTop 5 iPhone models by sales:")
print(iphone.sort_values("Sales", ascending=False).head())
plt.figure(figsize=(8,5))
plt.bar(iphone["Model"], iphone["Sales"])
plt.title("iPhone Sales by Model")
plt.xlabel("Model")
plt.ylabel("Sales (in millions)")
plt.show()


#  Project 2: Student Marks Analysis

print("\n--- Project 2: Student Marks Analysis ---")
students = pd.DataFrame({
    "Name": ["Aarav", "Ishita", "Rohan", "Meera", "Kabir",
             "Simran", "Arjun", "Priya", "Neha", "Kunal"],
    "Math": [78, 90, 65, 88, 95, 70, 82, 60, 85, 75],
    "Science": [85, 88, 70, 76, 89, 72, 79, 65, 92, 80],
    "English": [80, 92, 68, 85, 94, 74, 88, 70, 87, 72]
})
print(students.head())
students["Average"] = students[["Math", "Science", "English"]].mean(axis=1)
print("\nStudent averages:")
print(students[["Name", "Average"]])
topper = students.loc[students["Average"].idxmax()]
print(f"\nTopper: {topper['Name']} with average {topper['Average']}")
students.set_index("Name")[["Math", "Science", "English"]].plot(kind="bar", figsize=(8,5))
plt.title("Student Marks")
plt.ylabel("Marks")
plt.show()


#  Project 3: Netflix Movies Dataset Analysis

print("\n--- Project 3: Netflix Movies Dataset Analysis ---")
netflix = pd.DataFrame({
    "Title": ["Inception", "Interstellar", "The Dark Knight", "Parasite",
              "Avengers: Endgame", "Titanic", "The Irishman", "Joker",
              "Stranger Things", "Money Heist"],
    "Genre": ["Sci-Fi", "Sci-Fi", "Action", "Drama", "Action",
              "Romance", "Drama", "Drama", "Sci-Fi", "Action"],
    "Year": [2010, 2014, 2008, 2019, 2019, 1997, 2019, 2019, 2016, 2017],
    "Rating": [8.8, 8.6, 9.0, 8.6, 8.4, 7.9, 7.8, 8.5, 8.7, 8.3]
})
print(netflix.head())
genre_count = netflix["Genre"].value_counts()
print("\nMovies per Genre:")
print(genre_count)
genre_count.plot(kind="bar", figsize=(8,5), title="Number of Movies by Genre")
plt.show()
yearly = netflix["Year"].value_counts().sort_index()
yearly.plot(kind="line", figsize=(8,5), title="Movies Released per Year")
plt.show()
top10 = netflix.sort_values("Rating", ascending=False).head(10)
print("\nTop 10 Rated Movies:")
print(top10[["Title", "Genre", "Rating"]])
```
